## Data Augmentation
数据增强，由于数据集数量有限，对这些数据进行翻转，旋转，裁剪，加噪声等处理

```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
path = './extra_data/baiduget/'
savedpath = './extra_data/WM_Slime/'
image_width = 96
image_heigh = 96
```

```python
def img_change (path,savedpath) : 
    filelist = os.listdir(path)
    count = 36739
    for item in filelist :
        print(item)
        im = Image.open(path + item) #打开图片
        out = im.resize((image_width,image_heigh), Image.ADAPTIVE)
        
        item = item.replace('.jfif','.jpg')
        
        count=count + 1
        out = out.convert('RGB')
        out.save(savedpath + str(count)+'.jpg')
    #filelist = os.listdir(savedpath)    
    #for item in filelist :
        img = Image.open(savedpath + str(count)+'.jpg')
        img_ud = Image.fromarray(np.flipud(img)) # 上下翻转
        count=count + 1
        img_ud.save(savedpath + str(count)+'.jpg')
        
        img_lr = Image.fromarray(np.fliplr(img)) #左右翻转
        count=count + 1
        img_lr.save(savedpath + str(count)+'.jpg')
        
        img_R90 = Image.fromarray(np.rot90(img, 1)) # 顺时针旋转90度
        count=count + 1
        img_R90.save(savedpath + str(count)+'.jpg')
        
        img_L90 = Image.fromarray(np.rot90(img, -1)) # 顺时针旋转90度
        count=count + 1
        img_L90.save(savedpath + str(count)+'.jpg')
        
        img1 = np.array(img)
        noise = np.random.normal(0, 0.05, img1.shape)# 添加高斯噪声
        img2 = img1+noise
        img_noise = Image.fromarray(np.uint8(img2))
        count=count + 1
        img_noise.save(savedpath + str(count)+'.jpg')
        
        #print('item of %s is saved '%(item))
```

```python
img_change (path,savedpath) 
print('finish!')
```
